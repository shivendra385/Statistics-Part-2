
1. What is hypothesis testing in statistics
 ->Hypothesis testing is a statistical technique used to evaluate a hypothesis about a population parameter based on a sample of data.
2. What is the null hypothesis, and how does it differ from the alternative hypothesis
 ->Null and Alternative Hypotheses
 - Null Hypothesis (H0): A statement of no effect or no difference, which is tested for possible rejection.
 - Alternative Hypothesis (H1): A statement of an effect or difference, which is accepted if the null hypothesis is rejected.
3. What is the significance level in hypothesis testing, and why is it important*
 ->The significance level (α) is the maximum probability of rejecting the null hypothesis when it is true. It's a threshold for determining whether the observed effect is statistically significant.
4. What does a P-value represent in hypothesis testing*
 ->The P-value represents the probability of observing an effect as extreme or more extreme than the one observed, assuming the null hypothesis is true.
5.  How do you interpret the P-value in hypothesis testing
 ->If the P-value is below the significance level (α), the null hypothesis is rejected, indicating a statistically significant effect.
6.  What are Type 1 and Type 2 errors in hypothesis testing
 ->Type 1 and Type 2 Errors
- Type 1 Error: Rejecting the null hypothesis when it is true (false positive).
- Type 2 Error: Failing to reject the null hypothesis when it is false (false negative).
7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing
 ->One-Tailed and Two-Tailed Tests
- One-Tailed Test: Tests for an effect in one direction only.
- Two-Tailed Test: Tests for an effect in both directions.
8. What is the Z-test, and when is it used in hypothesis testing
 ->The Z-test is a statistical test used to compare the mean of a sample to a known population mean, or to compare the means of two samples.
9. How do you calculate the Z-score, and what does it represent in hypothesis testing
 ->Z = (X - μ) / σ, where X is the sample mean, μ is the population mean, and σ is the population standard deviation.
10. What is the T-distribution, and when should it be used instead of the normal distribution
 ->The T-distribution is a probability distribution used to estimate population parameters when the sample size is small.
11. What is the difference between a Z-test and a T-test
 ->Z-Test vs. T-Test
  Z-Test: Used when the population standard deviation is known.
- T-Test: Used when the population standard deviation is unknown.  
12. What is the T-test, and how is it used in hypothesis testing
  ->The T-test is a statistical test used to compare the means of two groups, or to compare the mean of a sample to a known population mean. It's commonly used in hypothesis testing to determine whether there's a significant difference between the means of two groups.
13.  What is the relationship between Z-test and T-test in hypothesis testing
 ->Relationship Between Z-Test and T-Test
The Z-test and T-test are both used for hypothesis testing, but they differ in their assumptions and applications:
a. Z-Test: Used when the population standard deviation is known, and the sample size is large (n > 30).
b. T-Test: Used when the population standard deviation is unknown, or the sample size is small (n ≤ 30).
14.  What is a confidence interval, and how is it used to interpret statistical results
 ->A confidence interval is a range of values within which a population parameter is likely to lie, with a certain level of confidence.
15. What is the margin of error, and how does it affect the confidence interval
 ->The margin of error is the maximum amount by which the sample mean may differ from the true population mean.
16. How is Bayes' Theorem used in statistics, and what is its significance
 ->Bayes' Theorem is a mathematical formula for updating the probability of a hypothesis based on new evidence.
17.  What is the Chi-square distribution, and when is it used
 ->The Chi-Square distribution is a probability distribution used to test hypotheses about categorical data.
18. What is the Chi-square goodness of fit test, and how is it applied
 ->The Chi-Square goodness of fit test is used to determine whether observed frequencies differ significantly from expected frequencies.
19. What is the F-distribution, and when is it used in hypothesis testing
 ->The F-distribution is a probability distribution used to test hypotheses about the variances of two populations.
20. What is an ANOVA test, and what are its assumptions

 ->Types of ANOVA Tests
- One-Way ANOVA: Compares the means of three or more groups.
- Two-Way ANOVA: Compares the means of two or more groups while controlling for the effect of a second variable.
22. What is the F-test, and how does it relate to hypothesis testing

```python
# 1Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
import numpy as np
from scipy.stats import norm

def z_test(sample_mean, population_mean, sample_std, sample_size):
    z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))
    p_value = 2 * (1 - norm.cdf(np.abs(z_score)))
    return z_score, p_value

sample_mean = 25
population_mean = 20
sample_std = 5
sample_size = 36

z_score, p_value = z_test(sample_mean, population_mean, sample_std, sample_size)
print("Z-Score:", z_score)
print("P-Value:", p_value)

```

```python
# 3  Implement a one-sample Z-test using Python to compare the sample mean with the population mean

import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample_data, population_mean):
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)
    sample_size = len(sample_data)

    z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))
    p_value = 2 * (1 - norm.cdf(np.abs(z_score)))

    return z_score, p_value

np.random.seed(0)
population_mean = 20
population_std = 5
sample_size = 36

sample_data = np.random.normal(population_mean, population_std, sample_size)
z_score, p_value = one_sample_z_test(sample_data, population_mean)

print("Z-Score:", z_score)
print("P-Value:", p_value)
```

```python
#4  Perform a two-tailed Z-test using Python and visualize the decision region on a plot

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test(sample_mean, population_mean, sample_std, sample_size):
    z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))
    p_value = 2 * (1 - norm.cdf(np.abs(z_score)))
    return z_score, p_value

sample_mean = 25
population_mean = 20
sample_std = 5
sample_size = 36

z_score, p_value = two_tailed_z_test(sample_mean, population_mean, sample_std, sample_size)

x = np.linspace(-3, 3, 100)
y = norm.pdf(x, loc=0, scale=1)
plt.plot(x, y)
plt.axvline(z_score, color='r', linestyle='--')
plt.axvline(-z_score, color='r', linestyle='--')
plt.title("Two-Tailed Z-Test Decision Region")
plt.xlabel("Z-Score")
plt.ylabel("Probability Density")
plt.show()
```

```python
# 5Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_type_1_and_type_2_errors(alpha, beta, mu0, mu1, sigma):
    z_alpha = norm.ppf(1 - alpha / 2)
    z_beta = norm.ppf(beta)

    type_1_error = norm.cdf(z_alpha, loc=mu0, scale=sigma)
    type_2_error = 1 - norm.cdf(z_beta, loc=mu1, scale=sigma)

    x = np.linspace(-3, 3, 100)
    y = norm.pdf(x, loc=mu0, scale=sigma)
    plt.plot(x, y)
    plt.axvline(z_alpha, color='r', linestyle='--')
    plt.title("Type 1 Error")
    plt.xlabel("Z-Score")
    plt.ylabel("Probability Density")
    plt.show()

    x = np.linspace(-3, 3, 100)
    y = norm.pdf(x, loc=mu1, scale=sigma)
    plt.plot(x, y)
    plt.axvline(z_beta, color='r', linestyle='--')
    plt.title("Type 2 Error")
    plt.xlabel("Z-Score")
    plt.ylabel("Probability Density")
    plt.show()

    return type_1_error, type_2_error

alpha = 0.05
beta = 0.2
mu0 = 20
mu1 = 25
sigma = 5

type_1_error, type_2_error = calculate_type_1_and_type_2_errors(alpha, beta, mu0, mu1, sigma)

print("Type 1 Error:", type_1_error)
print("Type 2 Error:", type_2_error)

```

```python
#9Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
from scipy.stats import norm

def calculate_confidence_interval(sample_data, confidence_level):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    sample_data (numpy array): Sample data.
    confidence_level (float): Confidence level (e.g., 0.95).

    Returns:
    tuple: Lower and upper bounds of the confidence interval.
    """
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)
    sample_size = len(sample_data)


    standard_error = sample_std / np.sqrt(sample_size)


    z_score = norm.ppf(1 - (1 - confidence_level) / 2)

    margin_of_error = z_score * standard_error



    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound
np.random.seed(0)
sample_data = np.random.normal(20, 5, 36)
confidence_level = 0.95

lower_bound, upper_bound = calculate_confidence_interval(sample_data, confidence_level)

print("Confidence Interval: ({}, {})".format(lower_bound, upper_bound))
```

```python
#10  Write a Python program to calculate the margin of error for a given confidence level using sample data

import numpy as np
from scipy.stats import norm

def calculate_margin_of_error(sample_data, confidence_level):
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)
    sample_size = len(sample_data)

    z_score = norm.ppf(1 - (1 - confidence_level) / 2)
    margin_of_error = z_score * (sample_std / np.sqrt(sample_size))

    return margin_of_error

np.random.seed(0)
sample_data = np.random.normal(20, 5, 36)
confidence_level = 0.95

margin_of_error = calculate_margin_of_error(sample_data, confidence_level)
print("Margin of Error:", margin_of_error)
```

```python
#11 Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

import numpy as np

def bayesian_inference(prior_probability, likelihood, evidence):
    posterior_probability = (prior_probability * likelihood) / evidence
    return posterior_probability

prior_probability = 0.5  # Prior probability of the hypothesis
likelihood = 0.8  # Likelihood of the data given the hypothesis
evidence = 0.6  # Evidence (normalizing constant)

posterior_probability = bayesian_inference(prior_probability, likelihood, evidence)
print("Posterior Probability:", posterior_probability)

```

```python
#13 Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data
import numpy as np

def calculate_expected_frequencies(observed_frequencies):
    row_totals = np.sum(observed_frequencies, axis=1)
    column_totals = np.sum(observed_frequencies, axis=0)
    total = np.sum(observed_frequencies)

    expected_frequencies = np.zeros_like(observed_frequencies)
    for i in range(observed_frequencies.shape[0]):
        for j in range(observed_frequencies.shape[1]):
            expected_frequencies[i, j] = (row_totals[i] * column_totals[j]) / total

    return expected_frequencies


observed_frequencies = np.array([[20, 10], [15, 25]])

expected_frequencies = calculate_expected_frequencies(observed_frequencies)
print("Expected Frequencies:\n", expected_frequencies)
```

```python
#16  Implement an F-test using Python to compare the variances of two random samples
import numpy as np
from scipy.stats import f

np.random.seed(0)
sample_size1 = 36
sample_size2 = 36
population_variance1 = 25
population_variance2 = 36

sample_data1 = np.random.normal(0, np.sqrt(population_variance1), sample_size1)
sample_data2 = np.random.normal(0, np.sqrt(population_variance2), sample_size2)

f_statistic = (np.var(sample_data1, ddof=1) / np.var(sample_data2, ddof=1))
p_value = f.sf(f_statistic, sample_size1 - 1, sample_size2 - 1)

print("F-Statistic:", f_statistic)
print("P-Value:", p_value)
```

```python
#19 Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA
import numpy as np
from scipy.stats import shapiro, bartlett

def check_anova_assumptions(group_data):
    # Normality assumption
    for i, group in enumerate(group_data):
        shapiro_statistic, shapiro_p_value = shapiro(group)
        print("Group {}: Shapiro Statistic={}, P-Value={}".format(i+1, shapiro_statistic, shapiro_p_value))

    # Equal variance assumption
    bartlett_statistic, bartlett_p_value = bartlett(*group_data)
    print("Bartlett Statistic={}, P-Value={}".format(bartlett_statistic, bartlett_p_value))

np.random.seed(0)
group_size = 36
group_means = [20, 25, 30]
group_std = 5

group_data = [np.random.normal(group_mean, group_std, group_size) for group_mean in group_means]

check_anova_assumptions(group_data)
```

```python
#20 Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the result
import numpy as np
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed(0)
factor1_levels = [20, 25]
factor2_levels = [30, 35]
group_size = 36
group_std = 5

data = []
for factor1_level in factor1_levels:
    for factor2_level in factor2_levels:
        group_data = np.random.normal(factor1_level + factor2_level, group_std, group_size)
        data.extend([(factor1_level, factor2_level, value) for value in group_data])

data = np.array(data)
df = pd.DataFrame(data, columns=['Factor1', 'Factor2', 'Value'])

model = ols('Value ~ C(Factor1) + C(Factor2) + C(Factor1):C(Factor2)', data=df).fit()
anova_table = anova_lm(model, typ=2)

print(anova_table)

plt.figure(figsize=(10, 6))
for factor2_level in factor2_levels:
    factor1_values = df[df['Factor2'] == factor2_level]['Factor1']
    value_means = df[df['Factor2'] == factor2_level]['Value'].groupby(factor1_values).mean()
    plt.plot(factor1_values.unique(), value_means, label='Factor2={}'.format(factor2_level))
plt.title("Two-Way ANOVA Interaction")
plt.xlabel("Factor1")
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python
#25 Write a Python script to perform a Z-test for comparing proportions between two datasets or groups
import numpy as np
from scipy.stats import norm

def z_test_for_proportions(sample_size1, sample_proportion1, sample_size2, sample_proportion2):
    pooled_proportion = (sample_proportion1 * sample_size1 + sample_proportion2 * sample_size2) / (sample_size1 + sample_size2)
    standard_error = np.sqrt(pooled_proportion * (1 - pooled_proportion) * (1 / sample_size1 + 1 / sample_size2))
    z_score = (sample_proportion1 - sample_proportion2) / standard_error
    p_value = 2 * (1 - norm.cdf(np.abs(z_score)))
    return z_score, p_value

sample_size1 = 100
sample_proportion1 = 0.4
sample_size2 = 120
sample_proportion2 = 0.45

z_score, p_value = z_test_for_proportions(sample_size1, sample_proportion1, sample_size2, sample_proportion2)

print("Z-Score:", z_score)
print("P-Value:", p_value)
```
